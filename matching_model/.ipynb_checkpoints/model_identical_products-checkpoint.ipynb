```python
class ArcMarginProduct(nn.Module):
    r"""Implement of large margin arc distance: :
        Args:
            in_features: size of each input sample
            out_features: size of each output sample
            s: norm of input feature
            m: margin

            cos(theta + m)
        """
    def __init__(self, in_features, out_features, s=30.0, m=0.50, easy_margin=False):
        super(ArcMarginProduct, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.s = s
        self.m = m
        self.weight = Parameter(torch.FloatTensor(out_features, in_features))
        nn.init.xavier_uniform_(self.weight)

        self.easy_margin = easy_margin
        self.cos_m = math.cos(m)
        self.sin_m = math.sin(m)
        self.th = math.cos(math.pi - m)
        self.mm = math.sin(math.pi - m) * m
        

    def forward(self, input, label):
        # --------------------------- cos(theta) & phi(theta) ---------------------------
        cosine = F.linear(F.normalize(input), F.normalize(self.weight))
        sine = torch.sqrt((1.0 - torch.pow(cosine, 2)).clamp(0, 1))
        phi = cosine * self.cos_m - sine * self.sin_m
        if self.easy_margin:
            phi = torch.where(cosine > 0, phi, cosine)
        else:
            phi = torch.where(cosine > self.th, phi, cosine - self.mm)
        # --------------------------- convert label to one-hot ---------------------------
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        one_hot = torch.zeros(cosine.size(), requires_grad=True, device=device)
        # -------------torch.where(out_i = {x_i if condition_i else y_i) -------------
        output = (one_hot * phi) + ((1.0 - one_hot) * cosine)  # you can use torch.where if your torch.__version__ is 0.4
        output *= self.s
        
        return output
```

```python
class TinyBert(nn.Module):
    '''Implement embedding from text
    Args:
        text(string): name or attributes
    Returns:
        embeddings(numpy.ndarray): embeddings, shape(batch, seq, emb)
    '''

    def __init__(self):
        super().__init__()
        self.m = BertForSequenceClassification.from_pretrained("cointegrated/rubert-tiny2")

    def forward(self, text):
        # на вход подаётся уже токенизированный текст

        with torch.no_grad():
            model_output = self.m(**{k: v.to(self.m.device) for k, v in text.items()}, output_hidden_states=True)

        """
        embeddings содержит 4 ембеддинга из разных слоёв. Стоит попробовать использовать каждый по отдельности, либо же MaxPooling, AveragePooling, GlobalPooling. нужно тестить , пока оставлю последний слой

        каждый такой эмбеддинг содержит ещё 240 слоёв, тоже можно попробовать пулинг
        """
        embeddings = model_output.hidden_states[-1][:, 0, :]
        embeddings = torch.nn.functional.normalize(embeddings)
        return embeddings.detach()
```

```python
class CNN(nn.Module):
    '''Implement embedding from img
    Args:

    Returns:

    '''

    def __init__(self, *args, **kwargs) -> None:
        super().__init__(*args, **kwargs)
        self.resnet = models.resnet34(pretrained=True)

    def forward(self, img):
        embedding = self.get_embedding(img)

        # возвращаем ембеддинг
        return embedding



    def get_embedding(self, img_tensor):
        img_tensor = img_tensor.to(device)
        layer = self.resnet._modules.get('avgpool')
        outputs = []

        def copy_embeddings(m, i, o):
            """Copy embeddings from the penultimate layer.
            """
            o = o[:, :, 0, 0].cpu().detach().numpy().tolist()
            outputs.append(o)
            hook.remove()

        # attach hook to the penulimate layer
        hook = layer.register_forward_hook(copy_embeddings)


        self.resnet.eval()

        self.resnet(img_tensor)
        list_embeddings = [item for sublist in outputs for item in sublist]
        self.embedding = np.array(list_embeddings)
        return self.embedding
```

```python
class Embedding(nn.Module):
    def __init__(self):
        super().__init__()


    def forward(self, name_bert, cnn, name, img, num_features, device): #img
        '''
        Args:
            name_bert: экземпляр TinyBert
            attr_bert: экземпляр TinyBert
            cnn: экземпляр CNN
            name: токенизированное название
            attr: токенизированные атрибуты
            img: image из загрузчика трансформированная
            num_features: количество фичей в финальном эмбеддинге
            device: экземпляр девайса cpu или cudo

        Returns:
            embedding: torch.tensor   size = (10, 1136)
        '''


        name = name_bert(name)

        img = torch.from_numpy(cnn(img)).to(device)


        self.BN = BatchNorm1d(num_features)
        embedding_raw = torch.cat((name, img), dim=1)
        self.BN.to(device, dtype=float)
        embedding = self.BN(embedding_raw.to(device, dtype=float))


        return embedding
```

### Main loss

```python
class CustomLoss(nn.Module):
    def __init__(self):
        super(CustomLoss, self).__init__()
        self.alpha = lambda x: 0.5 / np.exp(x/np.exp(1))  # [0, 0.5]?? or [0, 1]


    def forward(self, cnn_loss, names_bert_loss, arc_loss, epoch):

        main_loss = self.alpha(epoch)*(cnn_loss + names_bert_loss) + (1 - self.alpha(epoch)) * arc_loss
        return main_loss
```

### Visualization

```python
def umap_vector_visualization(dataframe:pd.DataFrame, show=None, color=None, dot_size=3, title=None, text=None, write_to_html_path=None):

    reducer = UMAP(n_components=3)
    vector = reducer.fit_transform(dataframe) # понижаем размерность
    fig = px.scatter_3d(x=vector[:, 0],
                        y=vector[:, 1],
                        z=vector[:, 2],
                        color=color,
                        title=title,
                        text=text
                        )
    fig.update_traces(marker=dict(size=dot_size))  # задаём размер точек

    if write_to_html_path:
        fig.write_html(write_to_html_path)   # если передан путь к html, файл будет записан в него

    if show:
        fig.show()  # отображение в  jupiter notebook
```

```python
# hyperparameters
num_epoch = 60
learning_rate = 1e-4
momentum = 0.9
batch_size = 10

# arcface
s = 30
m = 0.8

losses = [[], [], [], [], []]
losses_val = [[], [], [], [], []]
Loss_history = [[], [], [], [], []]
Loss_history_val = [[], [], [], [], []]

acc = [[], [], [], [], []]
acc5 = [[], [], [], [], []]
acc_val = [[], [], [], [], []]
Accuracy_history = [[], [], [], [], []]
Accuracy_history_val = [[], [], [], [], []]
```

```python
# инициализация
cnn = CNN()
names_bert = TinyBert()
attr_bert = TinyBert()
emb = Embedding()
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny2")
main_loss = CustomLoss()


'''# создаём загрузчики
train_loader = torch.utils.data.DataLoader(custom_dataset_train, batch_size=batch_size, shuffle=False)
val_loader = torch.utils.data.DataLoader(custom_dataset_val, batch_size=batch_size, shuffle=False)'''


# Заморозка всех слоев, кроме последнего
for param in cnn.resnet.parameters():
    param.requires_grad = False

for param in names_bert.m.parameters():
    param.requires_grad = False

for param in attr_bert.m.parameters():
    param.requires_grad = False

    
# заменяем последний слой
cnn.resnet.fc = torch.nn.Linear(cnn.resnet.fc.in_features, cat1_num_classes) 
names_bert.m.classifier = torch.nn.Linear(names_bert.m.classifier.in_features, cat1_num_classes)
attr_bert.m.classifier = torch.nn.Linear(attr_bert.m.classifier.in_features, cat1_num_classes)


# лоссы
cnn_criterion = torch.nn.CrossEntropyLoss()
names_bert_criterion = torch.nn.CrossEntropyLoss()
attr_bert_criterion = torch.nn.CrossEntropyLoss()


# оптимизаторы
optimizer1 = optim.SGD(cnn.resnet.fc.parameters(), lr=learning_rate)
optimizer2 = optim.Adam(names_bert.m.classifier.parameters(), lr=learning_rate)
optimizer3 = optim.Adam(attr_bert.m.classifier.parameters(), lr=learning_rate)
optimizer = optim.SGD([{'params': cnn.resnet.fc.parameters()},
                        {'params': names_bert.m.classifier.parameters()},
                        {'params': attr_bert.m.classifier.parameters()}],
                        lr=learning_rate)
scheduler = CosineAnnealingLR(optimizer, 1650*num_epoch)


final_embedding_lenght = cnn.resnet.fc.in_features + names_bert.m.classifier.in_features
metric_fc = ArcMarginProduct(final_embedding_lenght, cat1_num_classes, s=s, m=m) # cat1_num_classes or cat2_num_classes
arcface_loss = torch.nn.CrossEntropyLoss()


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
cnn.resnet = cnn.resnet.to(device)
names_bert.m = names_bert.m.to(device)
attr_bert.m = attr_bert.m.to(device)
metric_fc.to(device)


for epoch in range(num_epoch):
    cnn.resnet.train()
    names_bert.m.train()
    attr_bert.m.train()

    embbb_main = None
    embbb_attr = None
    cat1 = None
    cat2 = None
    names = None

    with open('loss.txt', 'a') as f:
        for i in trange(int(len(train) / batch_size)):
            torch.cuda.empty_cache()
            batch = train[i * batch_size : (i + 1) * batch_size]

            labels_cat1 = torch.tensor([p for p in batch['cat1_enc']])
            labels_cat2 = torch.tensor([p for p in batch['cat2_enc']])

            # токенизируем строки names и attributes
            names_train = tokenizer([p for p in batch['name']], padding=True, truncation=True, return_tensors='pt')
            attributes_train = tokenizer([p for p in batch['attributes']], padding=True, truncation=True, return_tensors='pt')

            # получаем изображения
            images_train = torch.stack([get_img(p).clone().detach().to(device) for p in batch['image']], dim=0)

            # обнуляем градиенты
            optimizer1.zero_grad()
            optimizer2.zero_grad()
            optimizer3.zero_grad()
            optimizer.zero_grad()

            # получаем предсказания
            cnn_outputs = cnn.resnet(images_train.to(device))
            names_bert_outputs = names_bert.m(**{k: v.to(device) for k, v in names_train.to(device).items()})
            attr_bert_outputs = attr_bert.m(**{k: v.to(device) for k, v in attributes_train.to(device).items()})

            # считаем лоссы на кат1
            cnn_loss = cnn_criterion(cnn_outputs, labels_cat1.to(device))
            names_bert_loss = names_bert_criterion(names_bert_outputs.logits, labels_cat1.to(device))
            attr_bert_loss = attr_bert_criterion(attr_bert_outputs.logits, labels_cat1.to(device))

            losses[0].append(cnn_loss.item())
            losses[1].append(names_bert_loss.item())
            losses[2].append(attr_bert_loss.item())

            # считаем acc@1 на кат1
            cnn_accuracy1 = top_k_accuracy_score(labels_cat1, cnn_outputs.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc'])))
            names_bert_accuracy1 = top_k_accuracy_score(labels_cat1, names_bert_outputs.logits.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc'])))
            attr_bert_accuracy1 = top_k_accuracy_score(labels_cat1, attr_bert_outputs.logits.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc'])))

            # считаем acc@5 на кат1
            cnn_accuracy5 = top_k_accuracy_score(labels_cat1, cnn_outputs.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc'])))
            names_bert_accuracy5 = top_k_accuracy_score(labels_cat1, names_bert_outputs.logits.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc'])))
            attr_bert_accuracy5 = top_k_accuracy_score(labels_cat1, attr_bert_outputs.logits.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc'])))

            acc[0].append(cnn_accuracy1.item())
            acc[1].append(names_bert_accuracy1.item())
            acc[2].append(attr_bert_accuracy1.item())

            acc5[0].append(cnn_accuracy5.item())
            acc5[1].append(names_bert_accuracy5.item())
            acc5[2].append(attr_bert_accuracy5.item())

            # получаем ембеддинг от трёх моделей
            embedding = emb(names_bert, cnn, names_train.to(device), images_train.to(device), final_embedding_lenght, device)
            embedding = embedding.detach()
            embedding.requires_grad = True

            # получаем предсказание кат2 от АркФейса
            output = metric_fc(embedding.float(), labels_cat1.to(device)) # cat1_num_classes or cat2_num_classes

            # считаем лосс по кат2
            arc_loss = arcface_loss(output, labels_cat1.to(device)) # labels_cat1 or labels_cat2

            # считаем acc@1 по кат2
            arc_accuracy1 = top_k_accuracy_score(labels_cat1, output.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc']))) # labels_cat1 or labels_cat2, list(set(data['cat1_enc'])) or list(set(data['cat2_enc']))

            # считаем acc@5 по кат2
            arc_accuracy5 = top_k_accuracy_score(labels_cat1, output.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc']))) # labels_cat1 or labels_cat2, list(set(data['cat1_enc'])) or list(set(data['cat2_enc']))

            # считаем финальный лосс
            loss = main_loss(cnn_loss, names_bert_loss, arc_loss, epoch)

            cnn_loss.backward(retain_graph=True)
            names_bert_loss.backward(retain_graph=True)
            attr_bert_loss.backward(retain_graph=True)
            loss.backward()

            losses[3].append(arc_loss.item())
            losses[4].append(loss.item())

            acc[3].append(arc_accuracy1.item())
            acc5[3].append(arc_accuracy5.item())

            optimizer1.step()
            optimizer2.step()
            optimizer3.step()
            optimizer.step()

            # записываем
            if (epoch+1) % 5 == 0 or epoch == 0:
                if embbb_main is None:
                    embbb_main = embedding.cpu().detach().numpy()
                else:
                    embbb_main = np.vstack((embbb_main, embedding.cpu().detach().numpy()))
                if embbb_attr is None:
                    embbb_attr = attr_bert_outputs.logits.cpu().detach().numpy()
                else:
                    embbb_attr = np.vstack((embbb_attr, attr_bert_outputs.logits.cpu().detach().numpy()))
                if cat1 is None:
                    cat1 = labels_cat1.cpu().detach().numpy()
                else:
                    cat1 = np.vstack((cat1, labels_cat1.cpu().detach().numpy()))
                if names is None:
                    names = batch['name_prepared'].values
                else:
                    names = np.vstack((names, batch['name_prepared'].values))
                    
            '''# val
            with torch.no_grad():
                indexes = random.sample(list(val.index), batch_size)

                labels_cat1 = torch.tensor([val['cat1_enc'][p] for p in sorted(indexes)]).type(torch.LongTensor)
                labels_cat2 = torch.tensor([val['cat2_enc'][p] for p in sorted(indexes)]).type(torch.LongTensor)

                # токенизируем строки names и attributes
                names_val = tokenizer([val['name'][p] for p in sorted(indexes)], padding=True, truncation=True, return_tensors='pt')
                attributes_val = tokenizer([val['attributes'][p] for p in sorted(indexes)], padding=True, truncation=True, return_tensors='pt')

                # получаем изображения
                images_val = torch.stack([get_img(val['image'][p]).clone().detach().to(device) for p in sorted(indexes)], dim=0)

                # получаем предсказания
                cnn_outputs = cnn.resnet(images_val.to(device))
                names_bert_outputs = names_bert.m(**{k: v.to(device) for k, v in names_val.to(device).items()})
                attr_bert_outputs = attr_bert.m(**{k: v.to(device) for k, v in attributes_val.to(device).items()})

                # считаем лоссы на кат1
                cnn_loss = cnn_criterion(cnn_outputs, labels_cat1.to(device))
                names_bert_loss = names_bert_criterion(names_bert_outputs.logits, labels_cat1.float().long().to(device))
                attr_bert_loss = attr_bert_criterion(attr_bert_outputs.logits, labels_cat1.float().long().to(device))

                losses_val[0].append(cnn_loss.detach().item())
                losses_val[1].append(names_bert_loss.detach().item())
                losses_val[2].append(attr_bert_loss.detach().item())

                # считаем acc@1 на кат1
                cnn_accuracy1 = top_k_accuracy_score(labels_cat1, cnn_outputs.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc'])))
                names_bert_accuracy1 = top_k_accuracy_score(labels_cat1.float().long(), names_bert_outputs.logits.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc'])))
                attr_bert_accuracy1 = top_k_accuracy_score(labels_cat1.float().long(), attr_bert_outputs.logits.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc'])))

                # считаем acc@5 на кат1
                cnn_accuracy5 = top_k_accuracy_score(labels_cat1, cnn_outputs.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc'])))
                names_bert_accuracy5 = top_k_accuracy_score(labels_cat1.float().long(), names_bert_outputs.logits.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc'])))
                attr_bert_accuracy5 = top_k_accuracy_score(labels_cat1.float().long(), attr_bert_outputs.logits.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc'])))

                acc_val[0].append(cnn_accuracy1.item())
                acc_val[1].append(names_bert_accuracy1.item())
                acc_val[2].append(attr_bert_accuracy1.item())

                # получаем ембеддинг от трёх моделей
                embedding = emb(names_bert, attr_bert, cnn, names_val.to(device), attributes_val.to(device), images_val.to(device), final_embedding_lenght, device)
                embedding = embedding.detach()
                embedding.requires_grad = True

                # получаем предсказание кат2 от АркФейса
                output = metric_fc(embedding.float(), labels_cat1.to(device)) # labels_cat1 or labels_cat2

                # считаем лосс по кат2
                arc_loss = arcface_loss(output, labels_cat1.float().long().to(device)) # labels_cat1 or labels_cat2

                # считаем финальный лосс
                loss = main_loss(cnn_loss, names_bert_loss, attr_bert_loss, arc_loss, epoch)

                losses_val[3].append(arc_loss.detach().item())
                losses_val[4].append(loss.detach().item())

                # считаем acc@1 по кат2
                arc_accuracy1 = top_k_accuracy_score(labels_cat1.float().long(), output.cpu().detach().numpy(), k = 1, labels = list(set(data['cat1_enc']))) # labels_cat1 or labels_cat2, list(set(data['cat1_enc'])) or list(set(data['cat2_enc']))
                acc_val[3].append(arc_accuracy1.item())

                # считаем acc@5 по кат2
                arc_accuracy5 = top_k_accuracy_score(labels_cat1.float().long(), output.cpu().detach().numpy(), k = 5, labels = list(set(data['cat1_enc']))) # labels_cat1 or labels_cat2, list(set(data['cat1_enc'])) or list(set(data['cat2_enc']))

                del (names_train, attributes_train, images_train, names_val, attributes_val, images_val, embedding)
                gc.collect()'''


        f.write(f"Epoch [{epoch}], img_loss: {mean(losses[0][-1844:])}, attr_loss: {mean(losses[2][-1844:])}, names_bert_loss: {mean(losses[1][-1844:])}, arc_loss: {mean(losses[3][-1844:])}, main_loss: {mean(losses[4][-1844:])}\n")
        f.write(f"Epoch [{epoch}], img_accuracy1: {mean(acc[0][-1844:])}, attr_accuracy1: {mean(acc[2][-1844:])}, names_bert_accuracy1: {mean(acc[1][-1844:])}, arc_accuracy1: {mean(acc[3][-1844:])}\n")
        f.write(f"Epoch [{epoch}], img_accuracy5: {mean(acc5[0][-1844:])}, attr_accuracy5: {mean(acc5[2][-1844:])}, names_bert_accuracy5: {mean(acc5[1][-1844:])}, arc_accuracy5: {mean(acc5[3][-1844:])}\n\n")

    print(f"Epoch [{epoch}], img_loss: {mean(losses[0][-1844:])}, attr_loss: {mean(losses[2][-1844:])}, names_bert_loss: {mean(losses[1][-1844:])}, arc_loss: {mean(losses[3][-1844:])}, main_loss: {mean(losses[4][-1844:])}\n")
    print(f"Epoch [{epoch}], img_accuracy1: {mean(acc[0][-1844:])}, attr_accuracy1: {mean(acc[2][-1844:])}, names_bert_accuracy1: {mean(acc[1][-1844:])}, arc_accuracy1: {mean(acc[3][-1844:])}\n")
    print(f"Epoch [{epoch}], img_accuracy5: {mean(acc5[0][-1844:])}, attr_accuracy5: {mean(acc5[2][-1844:])}, names_bert_accuracy5: {mean(acc5[1][-1844:])}, arc_accuracy5: {mean(acc5[3][-1844:])}\n\n")
    if (epoch+1) % 5 == 0 or epoch == 0:
        save_everything(epoch, names, cat1, embbb_main, embbb_attr) # + cat2

# saving models
torch.save(cnn.resnet.state_dict(), './data/models/resnet34.pth')
torch.save(names_bert.m.state_dict(), './data/models/names_bert.pth')
torch.save(attr_bert.m.state_dict(), './data/models/attr_bert.pth')
        
'''# colab     
torch.save(cnn.resnet.state_dict(), COLAB + './models/resnet34.pth')
torch.save(names_bert.m.state_dict(), COLAB + './models/names_bert.pth')
torch.save(attr_bert.m.state_dict(), COLAB + './models/attr_bert.pth')'''
```
